**实验目的**：验证 `ingestion/extractors/pdf.py` 的 `extract_to_markdown()` 函数

**验证样本**：

/experience_data/2020 - Learning Agile Robotic Locomotion Skills by Imitating Animals.pdf

/experience_data/2026年第二十五届全国大学生机器人大赛ROBOCON仿生足式机器人挑战赛比赛规则-V2.0.pdf

**验证事项**：
1. 函数能否正常调用、无报错
2. 双栏论文的文字顺序是否正确
3. 疑似标题行是否被加上 `##` 前缀

**验证结果**
1. 有些地方不该换行，比如一个段落被分隔两页，这里不该换行却被换行了；
2. 标题识别非常不准确，太多地方不是标题，却被识别成标题了


```python
import sys
sys.path.insert(0, '..')

from ingestion.extractors.pdf import extract_to_markdown, make_doc_id
print('导入成功')
```

```python
# 样本：学术论文（双栏、英文）
PDF_A = '../experience_data/2020 - Learning Agile Robotic Locomotion Skills by Imitating Animals.pdf'

md_a = extract_to_markdown(PDF_A)
print(f'doc_id : {make_doc_id(PDF_A)}')
print(f'总字符数: {len(md_a)}')
print(f'总行数  : {len(md_a.splitlines())}')
print(f'## 标题数: {md_a.count(chr(10) + "## ")}')

# 查看前面部分
for i, line in enumerate(md_a.splitlines()[:50]):
    print(f'{i+1:3d} | {line}')
```

```python
# 列出所有被识别为标题的行
headings_a = [line for line in md_a.splitlines() if line.startswith('## ')]
print(f'共识别出 {len(headings_a)} 个标题：')
for h in headings_a:
    print(' ', h)
```

```python
# 样本：比赛规则（单栏、中文）
PDF_B = '../experience_data/2026年第二十五届全国大学生机器人大赛ROBOCON仿生足式机器人挑战赛比赛规则-V2.0.pdf'

md_b = extract_to_markdown(PDF_B)
print(f'doc_id : {make_doc_id(PDF_B)}')
print(f'总字符数: {len(md_b)}')
print(f'总行数  : {len(md_b.splitlines())}')
print(f'## 标题数: {md_b.count(chr(10) + "## ")}')

# 查看前 60 行
for i, line in enumerate(md_b.splitlines()[:60]):
    print(f'{i+1:3d} | {line}')
```
